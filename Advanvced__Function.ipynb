```python
# Anonymous Function / Lambda Function
    # No Name.It is a special type of function wihout having the function name
# Lamnda function doesn't use the keyword  or statement called def
# inplace of the it use the lambda keyword


# Syntax  of lambda function:-
# lambda argument(s) : expression

# arguments : any value passed to this lambda function
# expression : conditions/logic to be executed


# lambda function with expression
greetings = lambda : print("Hello Everyone")


greetings()



# lambda function with arguments and expression

greet_user = lambda name : print("Hello there", name)

greet_user("Coachx")



batch_nos = lambda batch_Nos1 , batch_Nos2 : print("Your batch nos is " , batch_Nos1  , "and second is :" , batch_Nos2)

batch_nos(72 , 80)

batch_nos(72, 76)





```

```python
#Perform the addition , subtraction , Multiplication and Subtraction with the help of lambda function

add = lambda x , y : print(x+y , x-y , x*y , x/y)


add(4,5)



```

```python
# use of lambda with filter()

# syntax

# filter(function , list)

lst = [2,4,6,8,3,7,5,9,11,12,13,17,19,20,21]

lst_res = list(filter(

          lambda x : (x%2!=0) ,
          lst
))



list_res = lambda y : (y%2!=0)



print(lst_res)
print(list_res(4775785857))


```

```python
#use of map with reduce function

# syntax  of map :-

# map(function , lst)



list1 = [2,4,6,8,3,7,5,9,11,12,13,17,19,20,21]


list_result = list(map(

          lambda x : x*2 ,
          list1
))


print(list_result)




```

```python
#Use of reduce with lambda


# reduce :-
# It performs a repetitive operation over the pairs of the list
# syntax of reduce
# reduce(function , list):

l = [5,10,15,20,25,30]

# from module_name import function_name
from functools import reduce

Sum = reduce(
    lambda x , y : x+y,
    l
)

print(Sum)



```
